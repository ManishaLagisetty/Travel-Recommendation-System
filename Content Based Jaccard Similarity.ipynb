```python
# load  the datasett
df_business = pd.read_csv('/Users/sarahsha/Downloads/yelp_dataset_dataset_business_clean_withoriginalfeature.csv')
# create new binary columns based the level of restaurants price
df_dummies = pd.get_dummies(df_business['RestaurantsPriceRange2'], prefix='PriceRange')

# Concatenate the dummy variables with the original DataFrame
df_business = pd.concat([df_business, df_dummies], axis=1)

# Drop the original 'RestaurantsPriceRange2' column if needed
df_business = df_business.drop('RestaurantsPriceRange2', axis=1)
```

```python
# group the reviews by business_id
average_stars_per_business = filtered_review[['business_id','stars']].groupby('business_id')['stars'].mean().reset_index()
```

```python
# get the user's likes based on the rate of reviews, rate>=4 considered as likes
liked_business = average_stars_per_business[average_stars_per_business['stars'] >= 4]
```

```python
liked_business
```

```python
liked_business_list = liked_business['business_id'].tolist()
```

```python
# get all the business belong to restaurant category
df_restaurant = df_business[df_business['Restaurants'] == 1]
```

```python
# get the restaurants in the users' liked restaurants list 
liked_restaurant = df_restaurant[df_restaurant['business_id'].isin(liked_business_list)]
```

```python
liked_restaurant = liked_restaurant.reset_index(drop=True)
```

```python
liked_restaurant
```

```python
# get the user's favorite restuarant
liked_restaurant = df_restaurant[df_restaurant['business_id'].isin(liked_business_list)]
# select the feaures to calculate the Jaccard similarity
liked_restaurant = liked_restaurant[liked_restaurant['stars']>=4]
liked_restaurant = liked_restaurant.iloc[:, :-11].join(liked_restaurant.iloc[:, -5:])
liked_restaurant = liked_restaurant[liked_restaurant.select_dtypes(include='int').columns]
liked_restaurant = liked_restaurant.drop('review_count',axis=1)
```

```python
# get the candidate restaurants the users may be interested in
new_restaurants = df_restaurant[~df_restaurant['business_id'].isin(liked_business_list)]
new_restaurants = new_restaurants[new_restaurants['stars']>=4]
new_restaurants = new_restaurants.iloc[:, :-11].join(new_restaurants.iloc[:, -5:])
new_restaurants = new_restaurants[new_restaurants.select_dtypes(include='int').columns]
new_restaurants = new_restaurants.drop('review_count',axis=1)
```

```python
new_restaurants
```

```python
#import related library to calculate the jaccard_score
from sklearn.metrics import jaccard_score
for index_a, row_a in liked_restaurant.iterrows():
    most_similar = set()
    min_similarity = 0.9
    for index_b, row_b in new_restaurants.iterrows():
        similarity = round(jaccard_score(row_a, row_b, average='binary'),2)
        if similarity > min_similarity:
            most_similar.add(index_b)
        #print(f"Jaccard Similarity between Item {index_a} (A) and Item {index_b} (B): {similarity}")
print('the index of the restaurants with the jaccard similarity > 0.9')
print(most_similar)
```

```python
recommendation = df_business[['name','address','stars','review_count']].loc[list(most_similar)]
```
